Abrimos el archivo cargando unicamente las columnas necesarias para este ejercicio

```python
columnas_a_usar = ["App", "Rating", "Reviews"]
apps = pd.read_csv("googleplaystore.csv", usecols = columnas_a_usar)
```

```python
apps.count()
```

```python
apps.info()
```

Hay aplicaciones repetidas, las sacamos ya que en mi criterio no puede la misma app tener datos diferentes.

```python
apps["App"].duplicated().sum()
```

Nos quedamos unicamente con la primer repeticion de aquellas Apps que se repiten

```python
apps1 = apps.drop_duplicates(subset = "App")
apps1
```

Las eliminamos correctamente

```python
apps1["App"].duplicated().sum()
```

```python
apps1["Rating"].unique()
```

Hay un rating que no esta entre 1 y 5, lo eliminamos. Los ratings nans en este ejercicio los dejamos ya que no trabajamos mas con esta columna

```python
apps2 = apps1.drop(apps1[apps1["Rating"] > 5].index)
apps2["Rating"].unique()
```

Nos fijamos si hay reviews con valor nan

```python
apps2["Reviews"].isna().sum()
```

Convertimos la columna a int y comprobamos que todos sean numeros positivos

```python
apps2["Reviews"] = apps2["Reviews"].astype(int)
(apps2["Reviews"] >= 0).sum() == apps2["Reviews"].count()
```

Nos quedamos unicamente con la columna del nombre de la App de aquellas apps con al menos 1 review, las contamos

```python
apps3 = apps2.loc[apps2["Reviews"] >= 1]["App"]
apps3.count()
```
