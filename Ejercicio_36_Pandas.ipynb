```python
columnas_a_usar = ["App", "Rating", "Category", "Installs"]
apps = pd.read_csv("googleplaystore.csv", usecols = columnas_a_usar)
```

Hay aplicaciones repetidas, en mi criterio las sacamos ya que una misma app no puede tener datos diferentes.

```python
apps["App"].duplicated().sum()
```

Nos quedamos unicamente con la primer repeticion de aquellas Apps que se repiten

```python
apps1 = apps.drop_duplicates(subset = "App")
apps1
```

Los eliminamos correctamente

```python
apps1["App"].duplicated().sum()
```

```python
apps1["Rating"].unique()
```

Hay un rating que no esta entre 1 y 5, lo eliminamos. Los ratings nans en este ejercicio los dejamos ya que no trabajamos mas con esta columna

```python
apps2 = apps1.drop(apps1[apps1["Rating"] > 5].index)
apps2["Rating"].unique()
```

Para poder trabajar con enteros, le sacamos el '+' y las ',' a cada numero de la columna Installs y la convertimos en columna de enteros.

```python
apps2["Installs"] = apps2["Installs"].str.rstrip('+').replace(",", "", regex = True).astype(int)
apps2
```

```python
apps2.info()
```

Agrupamos por categoria y nos quedamos unicamente con las k+1 descargas de cada categoria ya que son las que vamos a utilizar. Probamos por ejemplo con k = 4

```python
k = 4
apps3 = apps2.groupby("Category")["Installs"].nlargest(k+1).reset_index()[["Category", "Installs"]]
apps3
```

Por ultimo, por categoria nos fijamos si la cantidad de descargas mas grande es mayor a la suma de las descargas de las k siguientes apps ordenadas por cantidad de descargas

```python
def es_dominante(categoria):
    primera_descarga = categoria.iloc[0]
    descargas_siguientes = categoria.iloc[1:].sum()
    if primera_descarga > descargas_siguientes:
        return "TIENE DOMINANTE"
    else:
        return "NO TIENE DOMINANTE"

apps4 = apps3.groupby("Category").agg(es_dominante)
apps5 = apps4.rename(columns = {"Installs" : "Dominancia"})
apps5
```
