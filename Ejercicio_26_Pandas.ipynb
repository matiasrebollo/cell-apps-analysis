```python
columnas_a_usar = ["App", "Rating", "Genres", "Reviews"]
apps = pd.read_csv("googleplaystore.csv", usecols = columnas_a_usar)
```

```python
apps
```

```python
apps.info()
```

Hay aplicaciones repetidas, en mi criterio las sacamos ya que una misma app no puede tener datos diferentes.

```python
apps["App"].duplicated().sum()
```

Nos quedamos unicamente con la primer repeticion de aquellas Apps que se repiten

```python
apps1 = apps.drop_duplicates(subset = "App")
apps1
```

Los eliminamos correctamente

```python
apps1["App"].duplicated().sum()
```

```python
apps1["Rating"].unique()
```

Hay un rating que no esta entre 1 y 5, lo eliminamos. Los ratings nans en este ejercicio los dejamos ya que no trabajamos mas con esta columna

```python
apps2 = apps1.drop(apps1[apps1["Rating"] > 5].index)
apps2["Rating"].unique()
```

Vemos que hay apps que tienen dos o mas generos. Son aquellas separadas por punto y coma. De paso revisamos que ninguna app tenga dos veces el mismo genero

```python
apps2["Genres"].unique()
```

Por lo que vamos a dejar separados esos generos pero que sigan perteneciendo a la misma app. Dejar los generos todos en minusculas nos asegura
que no haya generos iguales escritos de forma distinta, lo mismo con los espacios. ej: Beauty y beauty no es lo mismo.

```python
apps2["Genero"] = apps2.Genres.map(lambda x: [genero.lower().strip() for genero in x.split(";")])
apps3 = apps2.explode("Genero")
apps3
```

Nos fijamos que no hayan errores en los generos

```python
apps3["Genero"].unique()
```

Nos fijamos si hay reviews con valor nan

```python
apps3["Reviews"].isna().sum()
```

Convertimos la columna a int y comprobamos que todos sean numeros positivos

```python
apps3["Reviews"] = apps3["Reviews"].astype(int)
(apps3["Reviews"] >= 0).sum() == apps3["Reviews"].count()
```

Agrupamos por genero sumando las cantidades de reviews de cada genero. Al haber eliminado al comienzo las apps repetidas, nos aseguramos que
no figuren dos veces las reviews de una misma app. Es cierto que ahora aparecen apps repetidas, pero las reviews van para cada genero de esta app.
ej: Coloring book moana figura dos veces con la misma cantidad de reviews, pero son 967 para Art & Desing y 967 para Pretend Play.

```python
apps4 = apps3.groupby("Genero").sum()
apps4
```

Y por ultimo enseñamos el top 3 de generos con mayor cantidad de reviews

```python
apps4.nlargest(3, "Reviews")["Reviews"]
```
