Abrimos el archivo cargando unicamente las columnas necesarias para este ejercicio

```python
columnas_a_usar = ["App", "Rating", "Size"]
apps = pd.read_csv("googleplaystore.csv", usecols = columnas_a_usar)
```

```python
apps
```

```python
apps.info()
```

Hay aplicaciones repetidas, las sacamos ya que en mi criterio una misma app no puede tener datos diferentes

```python
apps["App"].duplicated().sum()
```

Nos quedamos unicamente con la primer repeticion de aquellas Apps que se repiten

```python
apps1 = apps.drop_duplicates(subset = "App")
apps1
```

Los eliminamos correctamente

```python
apps1["App"].duplicated().sum()
```

```python
apps1["Rating"].unique()
```

Hay un rating que no esta entre 1 y 5, lo eliminamos

```python
apps2 = apps1.drop(apps1[apps1["Rating"] > 5].index)
# Hay un rating que no esta entre 1 y 5, lo eliminamos
apps2["Rating"].unique()
```

```python
apps2["Size"].unique()
```

Nos quedamos unicamente con aquellas filas cuyo valor en "Size" sea "Varies with device"

```python
apps3 = apps2.loc[apps2["Size"] == "Varies with device"]
apps3
```

Vemos que hay algunos ratings que son NaNs, aca podemos tener dos criterios: o convertir esos NaNs a cierto numero, o eliminar aquellas filas
con rating NaN. El criterio que voy a tomar será eliminarlos.

```python
apps3["Rating"].isna().sum()
```

Las eliminamos correctamente

```python
apps4 = apps3.dropna(subset = "Rating")
apps4["Rating"].isna().sum()
```

Contamos cuantas apps tienen un tamaño que varia segun el dispositivo

```python
apps4["App"].count()
```

Y por ultimo enseñamos cual es la que tiene peor rating

```python
apps4.nsmallest(1, "Rating")
```
